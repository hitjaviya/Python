### 03) WAP to find out largest number from given two numbers using simple if and ternary operator.

```python
num1 = int(input("Please enter number 1 : "))
num2 = int(input("Please enter number 2 : "))

# Using simple if

if(num1>num2):
  print(f"{num1} is greater than {num2}")
else:
  print(f"{num2} is greater than {num1}")

# Using ternary operator  
print(f"{num1} is greater than {num2}") if(num1>num2) else print(f"{num2} is greater than {num1}")
```

### 05) WAP to check whether the given year is leap year or not.
[If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year = int(input("Please enter year in YYYY format : "))
if(year%4==0 and year%100!=0):
  print("Leap year")
else:
  print("Non leap year")  
```

### 01) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
Fail below 35 </br>
Pass Class between 35 to 45 </br>
Second Class</br>
between 45 to 60</br>
First Class between 60 to 70</br> 
Distinction if more than 70

```python
marks = []
result = 0
for x in range(5):
  marks.append(int(input(f"Please enter marks of subject {x+1} : ")))
  result = result + marks[x]
result = result/5
if(result>70):
  print("Result => Congratulations! You got Distinction")
elif(result>60 and result<=70):
  print("Result => Congratulations! You got First class")   
elif(result>45 and result<=60):
  print("Result => Congratulations! You got Second class")
elif(result>35 and result<=45):
  print("Result => Congratulations! You got Pass class")
else:
  print("Result => Fail")    
```

### 02) WAP to find out the Maximum and Minimum number from given 4 numbers.
