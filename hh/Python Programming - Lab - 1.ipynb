### 10) WAP to convert degree to Fahrenheit and vice versa.

```python
# Method -1 deciding on user input (using if condition)

choice = int(input("Please enter 1 for entering temperature in celsuis and 2 for enter 2 if temperature in Faherient : "))
temp   = int(input("Please enter the temperature : ")) 
if(choice == 1):
  print("Temperature in Faherient is : ",(temp*9/5)+32)
elif(choice == 2):
  print("Temperature in celsuis is : ",(temp-32)*5/9)
else:
  print("You entered wrong choice ")  

# Method -2 Separating both 

  # Celsuis to Faherient
temp1   = int(input("Please enter the temperature : ")) 
print("Temperature in Faherient is : ",(temp1*9/5)+32)

  # Faherient to Celsuis
temp2   = int(input("Please enter the temperature : "))   
print("Temperature in celsuis is : ",(temp2-32)*5/9)
```

### 12) Compute distance between two points taking input from the user (Pythagorean Theorem).

```python
X1 = int(input("Please enter the X1 co-ordinate : "))
Y1 = int(input("Please enter the Y1 co-ordinate : "))
X2 = int(input("Please enter the X2 co-ordinate : "))
Y2 = int(input("Please enter the Y2 co-ordinate : "))  
print("The distance between both the points is : ",round((((X2-X1)**2 + (Y2-Y1)**2)**0.5),2))
```

### 13) WAP to convert seconds into hours, minutes & seconds and print in HH:MM:SS
[e.g. 10000 seconds mean 2:46:40 (2 Hours, 46 Minutes, 40Seconds)]

```python
sec    = int(input("Please enter seconds : "))
hour   = int(sec/3600)
min    = int(sec/60)
if(min>=60):
  min = int(min%60)
s   = int(sec%60)
print(f"Time => {hour}:{min}:{s}")
```
